```python
# Function to capture photos for training data
def capture_photos(subject_name, output_folder):
    # Initialize camera
    camera = cv2.VideoCapture(0)
    cv2.namedWindow("Capture Photos")
    
    # Create directory if it doesn't exist
    output_path = os.path.join(output_folder, subject_name)
    os.makedirs(output_path, exist_ok=True)
    
    # Capture photos
    count = 0
    while True:
        ret, frame = camera.read()
        cv2.imshow("Capture Photos", frame)
        
        # Press 'q' to quit capturing
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
        
        # Press 's' to save the captured photo
        if cv2.waitKey(1) & 0xFF == ord('s'):
            count += 1
            img_name = os.path.join(output_path, "{}.jpg".format(count))
            cv2.imwrite(img_name, frame)
            print("{} saved!".format(img_name))  # Print confirmation message
            # Delay to ensure the message is displayed before closing window
            cv2.waitKey(1000)  
            # Delay to avoid capturing multiple images with a single press
            while cv2.waitKey(1) & 0xFF == ord('s'):  
                pass
    
    camera.release()
    cv2.destroyAllWindows()

# Example usage:
subject_name = input("Enter subject name: ")
output_folder = "/home/pmishra/Desktop/Projects/Github Projects/Face_Detect/Data"
capture_photos(subject_name, output_folder)


# # Function to detect face and save photos
# def capture_photos(subject_name, output_folder):
#     # Initialize camera
#     camera = cv2.VideoCapture(0)
#     cv2.namedWindow("Capture Photos")
    
#     # Create directory if it doesn't exist
#     output_path = os.path.join(output_folder, subject_name)
#     os.makedirs(output_path, exist_ok=True)
    
#     # Initialize face detector
#     face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

#     # Capture photos
#     count = 0
#     while count < 50:  # Capture up to 50 photos
#         ret, frame = camera.read()
#         if not ret:
#             print("Failed to capture image from camera!")
#             break
        
#         # Convert frame to grayscale for face detection
#         gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        
#         # Detect faces in the frame
#         faces = face_cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5)
        
#         # Draw rectangles around detected faces and save photos on 's' key press
#         for (x, y, w, h) in faces:
#             cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)
#             if cv2.waitKey(1) & 0xFF == ord('s'):
#                 count += 1
#                 img_name = os.path.join(output_path, "{}.jpg".format(count))
#                 cv2.imwrite(img_name, frame)
#                 print("{} saved!".format(img_name))  # Print confirmation message
        
#         # Display frame with rectangles drawn around faces
#         cv2.imshow("Capture Photos", frame)
        
#         # Press 'q' to quit capturing
#         if cv2.waitKey(1) & 0xFF == ord('q'):
#             break
    
#     # Release camera and close OpenCV windows
#     camera.release()
#     cv2.destroyAllWindows()

# # Example usage:
# subject_name = input("Enter subject name: ")
# output_folder = "/home/pmishra/Desktop/Projects/Github Projects/Face_Detect/Data"
# capture_photos(subject_name, output_folder)


```

```python
# Function to detect faces in an image
def detect_face(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.2, minNeighbors=5)
    
    if len(faces) == 0:
        return None, None
    
    (x, y, w, h) = faces[0]
    
    return gray[y:y+w, x:x+h], faces[0]
    
# Function to prepare training data
def prepare_training_data(data_folder):
    subjects = []
    faces = []
    labels = []
    
    # Iterate through each subject folder
    for subject_name in os.listdir(data_folder):
        subject_dir = os.path.join(data_folder, subject_name)
        
        # Skip files and folders that are not directories
        if not os.path.isdir(subject_dir):
            continue
        
        subjects.append(subject_name)
        label = len(subjects) - 1
        
        # Iterate through each image in the subject folder
        for image_name in os.listdir(subject_dir):
            image_path = os.path.join(subject_dir, image_name)
            
            # Read image
            image = cv2.imread(image_path)
            
            # Detect face
            face, rect = detect_face(image)
            
            if face is not None:
                faces.append(face)
                labels.append(label)
    
    cv2.destroyAllWindows()
    
    return faces, np.array(labels), subjects

# Example usage:
data_folder = "/home/pmishra/Desktop/Projects/Github Projects/Face_Detect/Data"
faces, labels, subjects = prepare_training_data(data_folder)
print(f"Total faces: {len(faces)}")
print(f"Total labels: {len(labels)}")
print(f"Subjects: {subjects}")
```

```python
# Function to prepare training data
def prepare_training_data(data_folder_path):
    # Get the directories (each directory contains images of one subject)
    directories = os.listdir(data_folder_path)
    faces = []
    labels = []
    
    # Traverse through each directory and read images within it
    for dir_name in directories:
        if not dir_name.startswith("subject"):
            continue
        
        # Extract label number from directory name
        label = int(dir_name.replace("subject", ""))
        
        # Path to the subject directory
        subject_dir_path = os.path.join(data_folder_path, dir_name)
        
        # Get the images names that are inside the subject directory
        subject_images_names = os.listdir(subject_dir_path)
        
        # Traverse through each image name, read image, and add to training list
        for image_name in subject_images_names:
            # Ignore system files like .DS_Store
            if image_name.startswith("."):
                continue
            
            # Path for the image
            image_path = os.path.join(subject_dir_path, image_name)
            
            # Read the image
            image = cv2.imread(image_path)
            
            # Convert image to grayscale
            gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
            
            # Add the face and label to the lists
            faces.append(gray_image)
            labels.append(label)
            
    return faces, labels

# Function to train the face recognition model
def train_face_recognizer(data_folder_path):
    # Prepare data
    faces, labels = prepare_training_data(data_folder_path)
    
    # Initialize LBPH face recognizer
    face_recognizer = cv2.face.LBPHFaceRecognizer_create()
    
    # Train the face recognizer
    face_recognizer.train(faces, np.array(labels))
    
    # Save the trained model to a file
    face_recognizer.save("face_recognizer.xml")
    
    print("Training completed. Model saved as 'face_recognizer.xml'.")

# Example usage:
data_folder_path = "/home/pmishra/Desktop/Projects/Github Projects/Face_Detect/Data"
train_face_recognizer(data_folder_path)
```

```python
# Function to detect face using OpenCV
def detect_face(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.2, minNeighbors=5)
    
    if len(faces) == 0:
        return None, None
    
    (x, y, w, h) = faces[0]
    return gray[y:y+w, x:x+h], faces[0]

```

```python
# Function to prepare training data
def prepare_training_data(data_folder_path):
    dirs = os.listdir(data_folder_path)
    faces = []
    labels = []
    
    for dir_name in dirs:
        if not dir_name.startswith("s"):
            continue
        label = int(dir_name.replace("s", ""))
        subject_dir_path = os.path.join(data_folder_path, dir_name)
        subject_images_names = os.listdir(subject_dir_path)
        
        for image_name in subject_images_names:
            if image_name.startswith("."):
                continue
            image_path = os.path.join(subject_dir_path, image_name)
            image = cv2.imread(image_path)
            face, rect = detect_face(image)
            
            if face is not None:
                faces.append(face)
                labels.append(label)
    
    return faces, labels
```
